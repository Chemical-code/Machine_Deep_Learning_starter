```python
keras.layers.Conv2D(10, kernal_size=(3,3), activation='relu', padding='valid')
```
#### same padding
```python
keras.layers.Conv2D(10, kernal_size=(3,3), activation='relu', padding='same')
```

```python
kears.layers.Conv2d(10, kernal_size=(3,3), actiavtion='relu', padding='same, strides=1)
```
